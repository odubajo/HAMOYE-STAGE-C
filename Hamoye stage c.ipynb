```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Data_for_UCI_named.csv')
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# convert the categorical target to binary
df["stabf"] = df["stabf"].replace(["unstable","stable"],[0,1])
df.head()
```

```python
df = df.drop("stab",axis = 1)
df.head()
```

```python
# Standiize input features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
for feature in df.columns:
  if feature != "stabf":
    df[feature] = scaler.fit_transform(df[[feature]])
df.head()
```

```python
# splitting data into training and testing
from sklearn.model_selection import train_test_split
X = df.drop("stabf",axis=1)
y = df["stabf"]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.3 , random_state= 101)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import ExtraTreesClassifier
extra_tree_forest = ExtraTreesClassifier(random_state=1)
parameters = {
    "max_depth": [2,3, 5, 10, None,"auto"],
    "n_estimators": [100, 200, 300, 400, 500,1000],
    "min_samples_split":[5,6,7,8],
    "min_samples_leaf":[5,6,7,8]
}
clf = RandomizedSearchCV( ExtraTreesClassifier(),parameters, cv=5, n_iter=10, scoring = 'accuracy', n_jobs = -1, verbose = 1, random_state = 1)
search = clf.fit(x_train,y_train)
search.best_params_
```

```python
extra_tree_new= ExtraTreesClassifier(n_estimators=300,min_samples_split=5,min_samples_leaf=6,max_depth=None)
from sklearn.metrics import accuracy_score
extra_tree_new.fit(x_train,y_train)
extra_tree_new_pred = extra_tree_new.predict(x_test)
# calculate accuracy score
Accuracy = accuracy_score(extra_tree_new_pred,y_test)
print(Accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=1)
rfc.fit(x_train,y_train)
rfc_preds = rfc.predict(x_test)
# calculate accuracy score
Accuracy = accuracy_score(rfc_preds,y_test)
print("Accuracy Score:",Accuracy)
```

```python
from xgboost import XGBClassifier
xgclass = XGBClassifier(random_state=1)
xgclass.fit(x_train,y_train)
xgb_pred = xgclass.predict(x_test)
# calculate accuracy score
Accuracy = accuracy_score(xgb_pred,y_test)
print(Accuracy)
```
